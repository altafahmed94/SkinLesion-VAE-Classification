```python
import pandas as pd

# Load the dataframe, treating the first column as the index
df = pd.read_csv('/kaggle/input/metadata-filled-86-5/filledValues_0.8652173913043478.csv', index_col=0)

df = df.drop(columns=['patient_id', 'lesion_id'])

print(df)
```

```python
import os
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow.keras.applications import DenseNet201
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.densenet import preprocess_input

# Initialize the DenseNet201 model
model = DenseNet201(weights='imagenet', include_top=False, pooling='avg')

# Define a function to extract features from an image
def extract_features(img_path):
    img = image.load_img(img_path, target_size=(1024, 1024))
    img_array = image.img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0)
    img_array = preprocess_input(img_array)
    features = model.predict(img_array)
    return features.flatten()  # Flatten to make it 1D array

feature_list = []
for index, rows in df.iterrows():
    image_path1 = f"/kaggle/input/hair-removed-pad-ufes/hair_removed_pad_ufes/{df.at[index, 'img_id']}"
    if os.path.exists(image_path1):
        features = extract_features(image_path1)
        feature_list.append(features)
        
# Create a DataFrame from the extracted features
features_df = pd.DataFrame(feature_list, columns=[f'feature_{i}' for i in range(features_df.shape[1])])

# Concatenate the original DataFrame with the features DataFrame
df = pd.concat([df, features_df], axis=1)

print(df.head())
```
